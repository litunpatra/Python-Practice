## "global" key word

When we want to **MODIFY the GLOBAL VARIABLE values** ___from within a function definition___, then ___the global variable names must be preceded with___ **'global' keyword** otherwise we get "**UnboundLocalError**: local variable names referenced before assignment"

```python
#globalvarex1.py
a=10
def access1():
    print("Val of a =",a)

#main
access1()
```

In the above example as we are just accessing the value of the global variable "a", thus it is not required to make use of the global keyword while accessing "a" from within the function named "access1()".

```python
#globalvarex2.py
a=10
def access1():
    global a
    a=a+5
    print("Inside access1() Value of \"a\" = {}".format(a))

#main():
print("Before execution of access1(), Value of \"a\" in main = {}".format(a))
access1()
print("After execution of access1(), Value of \"a\" in main = {}".format(a))
```

```python
#globalvarex3.py
def update1():
    global a,b
    a=a+10
    b=b+10

def update2():
    global a,b
    a=a*10
    b=b*10

#main
a,b=1,2
print("Before execution of update functions:")
print("Value of a={}\nb={}".format(a,b))
print("-"*50)

update1()
print("After update1():")
print("Value of a={}\nb={}".format(a,b))
print("-"*50)

update2()
print("After update2():")
print("Value of a={}\nb={}".format(a,b))
print("-"*50)
```

```python
#globalvarex4.py
def update1():
    a,b=10,20  #local variables with same name as global variables  
    a=a+10       
    b=b+10
    #local variables are updated above NOT Global Variables
    print("Values within update1():")
    print("a = {} \nb = {}".format(a,b))
    

def update2():
    a,b=5,7    #local variables with same name as global variables  
    a=a*10
    b=b*10
    #local variables are updated above NOT Global Variables
    print("Values within update2():")
    print("a = {} \nb = {}".format(a,b))
    
def update3():
    global a,b    
    a=a*10
    b=b*10
    #Global variables are updated above  
    print("Values within update3():")
    print("a = {} \nb = {}".format(a,b))

#main
a,b=100,200
print("Before execution of update functions:")
print("a = {}\nb = {}".format(a,b))
print("-"*50)

update1()
print("After update1():")
print("a = {}\nb = {}".format(a,b))
print("-"*50)

update2()
print("After update2():")
print("a = {}\nb = {}".format(a,b))
print("-"*50)

update3()
print("After update3():")
print("a = {}\nb = {}".format(a,b))
print("-"*50)
```

### global and local variables and globals()
- When "global" variable names are same with "local" variable names within a function definition, then PVM gives preference to local variables inside the scope of that function instead of the global variables.
- In such cases, to access the values of the global variables along with the values of the local variables , we use **globals()** 
  - it returns an object of <class, 'dict'> 
    - The above "dict" object stores all **global variable names** as **keys** and all corresponding **global variable values** as **values**. 

```python
#globalsfunex1.py
a,b,c,d = 10,20,30,40   # global variables
def operation():
    a,b,c,d = 1,2,3,4  # local variables
    print("="*50)
    print("Printing Values of Local Variables:")
    print("="*50)
    print("a = {}".format(a))
    print("b = {}".format(b))
    print("c = {}".format(c))
    print("d = {}".format(d))
    print("="*50)
    print()
    print("="*50)
    print("Printing Values of Global Variables:")
    print("="*50)
    print("a = {}".format(globals().get('a')))
    print("b = {}".format(globals().get('b')))
    print("c = {}".format(globals().get('c')))
    print("d = {}".format(globals().get('d')))
    print("="*50)

#main
operation()
```

```python
#globalsfunex2.py
a=10
b=20
c=30
d=40
def fun1():
    gv=globals()                           #gv belongs to <class, 'dict'>
    for k,v in gv.items():
        print("\t{}\t{}".format(k,v))
    print("-"*50)
    print("Accessing values of global variables from 'dict-obj' through indexing:")
    print("val of a = ",gv['a'])
    print("val of b = ",gv['b'])
    print("val of c = ",gv['c'])
    print("val of d = ",gv['d'])
    print("-"*50)
    print("Accessing values of global variables from 'dict-obj' through get():")
    print("val of a = ",gv.get('a'))
    print("val of b = ",gv.get('b'))
    print("val of c = ",gv.get('c'))
    print("val of d = ",gv.get('d'))
    print("-"*50)
    print("Accessing values of global variables from 'globals()' through get():")
    print("val of a = ",globals()['a'])
    print("val of b = ",globals()['b'])
    print("val of c = ",globals()['c'])
    print("val of d = ",globals()['d'])
    print("-"*50)
    print("Accessing values of global variables from 'globals()' through get():")
    print("val of a = ",globals().get('a'))
    print("val of b = ",globals().get('b'))
    print("val of c = ",globals().get('c'))
    print("val of d = ",globals().get('d'))
    print("-"*50)

#main:
fun1()
```

```python
#globalsfunex4.py
a=10
b=20
c=30
d=40

def oper():
    a,b,c,d = 1,2,3,4
    print("-"*50)
    print("Value of local - a = {}".format(a))
    print("Value of local - b = {}".format(b))
    print("Value of local - c = {}".format(c))
    print("Value of local - d = {}".format(d))
    print("-"*50)
    print("Value of global - a = {}".format(globals()['a']))
    print("Value of global - b = {}".format(globals()['b']))
    print("Value of global - c = {}".format(globals().get('c')))
    print("Value of global - d = {}".format(globals().get('d')))
    print("-"*50)
    res = a + b + c + d + globals()['a'] + globals()['b'] + \
            globals().get('c') + globals().get('d')
    print("Sum of values of all local & global variables = {}".format(res))
    print("-"*50)
    
#main
oper()
```
